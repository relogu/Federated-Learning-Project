```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import skewnorm, uniform
```

```python
N_SAMPLES = 2043
N_CLIENTS = 10
SEED = 51550
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 15}
matplotlib.rc('font', **font)
```

```python
r = uniform.rvs(size=N_SAMPLES, random_state=SEED)
fig, ax = plt.subplots(figsize=(16, 8))
histo = plt.hist(r, bins=N_CLIENTS, rwidth=0.8, align='left')
plt.xticks(ticks=histo[1][:-1],
           labels=['Client {}'.format(c) for c in np.arange(0,N_CLIENTS)])
plt.grid(axis='y')
plt.ylabel('number of samples')
plt.xlabel('clients')
plt.savefig('unif_distr_clients.png')
plt.show()
```

```python
BALANCE = 4
r = skewnorm.rvs(BALANCE, size=N_SAMPLES, random_state=SEED)
fig, ax = plt.subplots(figsize=(16, 8))
histo = plt.hist(r, bins=N_CLIENTS, rwidth=0.8, align='left')
plt.xticks(ticks=histo[1][:-1],
           labels=['Client {}'.format(c) for c in np.arange(0,N_CLIENTS)])
plt.grid(axis='y')
plt.ylabel('number of samples')
plt.xlabel('clients')
plt.savefig('skew_gauss_distr_clients.png')
plt.show()
```
