```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import pathlib
```

```python
PATH_TO_CSV = pathlib.Path('/home/relogu/Desktop/OneDrive/UNIBO/Magistrale/Federated Learning Project/ray_results')
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 18}
matplotlib.rc('font', **font)
```

What is best??
Here we want highest possible ``cos_silh_score``, ``data_calinski_harabasz``.
Then highest possible ``eucl_silh_score``, ``feat_calinski_harabasz``.
Then minimum ``cl_recon``.

```python
df = pd.DataFrame()
for arch in ['dec', 'google', 'curves']:
    for opt in ['adam', 'sgd', 'yogi']:
        tmp = pd.read_csv(PATH_TO_CSV/'euromds_cl_{}_{}.csv'. \
            format(arch, opt))
        tmp['arch'] = arch
        tmp['opt'] = opt
        df = df.append(tmp)
df.head()
```

```python
df.sort_values(
    axis=0,
    by=['ray/tune/cos_sil_score', 'ray/tune/data_calinski_harabasz', 'ray/tune/eucl_sil_score', 'ray/tune/feat_calinski_harabasz', 'ray/tune/cl_recon'],
    ascending=[False, False, False, False, True],
    inplace=True
    )
df
```

```python
df['ray/tune/data_calinski_harabasz'] = np.log(df['ray/tune/data_calinski_harabasz'].to_numpy())
df['ray/tune/feat_calinski_harabasz'] = np.log(df['ray/tune/feat_calinski_harabasz'].to_numpy())
df[:10].round(5).to_csv('euromds_cl_best.csv')
```

```python
for arch in ['dec', 'google', 'curves']:
    for opt in ['adam', 'sgd', 'yogi']:
        print(df[(df['arch']==arch)&(df['opt']==opt)].head())
```

```python
df['lr'] = np.log(df['lr'].to_numpy())
df['ray/tune/cl_recon'] = np.log(df['ray/tune/cl_recon'].to_numpy())
df['ray/tune/data_calinski_harabasz'] = np.log(df['ray/tune/data_calinski_harabasz'].to_numpy())
df['ray/tune/feat_calinski_harabasz'] = np.log(df['ray/tune/feat_calinski_harabasz'].to_numpy())
print(df.head())
figure, ax = plt.subplots(1, 1, figsize=(40,40))
tmp = df[~df.isin([np.nan, np.inf, -np.inf]).any(1)]
pd.plotting.scatter_matrix(tmp, ax=ax, grid=True)
plt.show()
```
